```python
# Importación de librerías necesarias
import pandas as pd
import os
from IPython.display import display

# Clase para gestionar y unificar DataFrames de DemandArea
class DemandAreaDataMerger:
    def __init__(self, folder_path):
        self.folder_path = folder_path
        self.dataframes = {}
        
    def load_dataframes(self):
        """Carga todos los archivos CSV de la carpeta especificada en dataframes."""
        try:
            files = [f for f in os.listdir(self.folder_path) if f.endswith('.csv')]
            for file in files:
                file_path = os.path.join(self.folder_path, file)
                df_name = file.replace('.csv', '')
                self.dataframes[df_name] = pd.read_csv(file_path)
                print(f"Cargado: {df_name} ({file})")
        except Exception as e:
            print(f"Error al cargar archivos: {e}")

    def merge_dataframes(self):
        """Une todos los DataFrames relevantes basándose en las claves comunes."""
        try:
            base_df = self.dataframes.get('das_config')
            if base_df is None:
                raise ValueError("'das_config' no encontrado. Es necesario para la base de unificación.")

            df_list = [k for k in self.dataframes.keys() if k != 'das_config']
            merged_df = base_df.copy()

            for df_name in df_list:
                merged_df = merged_df.merge(
                    self.dataframes[df_name],
                    on=["DemandArea_Id", "Region", "ElArea_Id", "LpgArea_Id", "Area_Type"],
                    how="left"
                )
                print(f"Unido: {df_name}")

            print("✅ Todos los DataFrames unificados exitosamente.")
            return merged_df

        except Exception as e:
            print(f"❌ Error durante la unificación: {e}")
```
